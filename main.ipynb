## Se iteran las regiones para acceder a los siguientes datos requeridos:
* city name
* language


## una vez obtenidos estos datos se llana una lista con el tiempo que tardo en procesar  cada dato, para posteriormente crear un dataframe

```python

bd = []
for region in regiones:
    assert(region)
    start_time = time.time()
    paises = obtener_pais_por_region(region)
    try:
        for pais in paises:
            start_time = time.time()
            for clave,valor in pais['languages'].items():
                valor = hashlib.sha1(valor.encode('UTF-8')).hexdigest()
                bd.append([region,pais['name']['common'],valor, time.time()-start_time])
            
    except Exception as e:

        print("no tiene lenguaje")

dataframe = pd.DataFrame(bd, columns=["Region", "City Name", "Language","Time"])
```

## Con funciones de la librería pandas se muestra lo siguiente
* El tiempo total que tardo en procesar todas las filas de la tabla
* El tiempo promedio que tardo en procesar todas las filas de la tabla
* El tiempo mínimo que tardo en procesar todas las filas de la tabla
* El máximo que tardo en procesar todas las filas de la tabla

```python
print(f"Tiempo total: {dataframe['Time'].sum()} s")
print(f"Tiempo promedio: {dataframe['Time'].mean()} s")
print(f"Tiempo minimo: {dataframe['Time'].min()} s")
print(f"Tiempo maximo: {dataframe['Time'].max()} s")
```
